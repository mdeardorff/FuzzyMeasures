```python
def stack_decl_jsons(jsons):
    base_json = jsons[0]
    num_frames = max([int(x[1:]) for x in base_json["frameDeclarations"].keys()]) + 1
    
    
    for jsonf in jsons[1:]:
        for i, f_decl in enumerate(jsonf["frameDeclarations"]):
            decl = jsonf["frameDeclarations"][str(f_decl)]["declarations"]
#             print(decl)
            base_json["frameDeclarations"]["f" + str(int(num_frames + i))] = {}
            base_json["frameDeclarations"]["f" + str(int(num_frames + i))]["declarations"] = []
            base_json["frameDeclarations"]["f" + str(int(num_frames + i))]["declarations"] = decl
        num_frames = num_frames + max([int(x[1:]) for x in jsonf["frameDeclarations"].keys()])+ 1
    
    return base_json

def stack_truth_jsons(jsons):
    base_json = jsons[0]
    num_frames = max([int(x[1:]) for x in base_json["frameAnnotations"].keys()]) + 1
    
    
    for jsonf in jsons[1:]:
        for i, f_decl in enumerate(jsonf["frameAnnotations"]):
            decl = jsonf["frameAnnotations"][str(f_decl)]["annotations"]
#             print(decl)
            base_json["frameAnnotations"]["f" + str(int(num_frames + i))] = {}
            base_json["frameAnnotations"]["f" + str(int(num_frames + i))]["annotations"] = []
            base_json["frameAnnotations"]["f" + str(int(num_frames + i))]["annotations"] = decl
        num_frames = num_frames + max([int(x[1:]) for x in jsonf["frameAnnotations"].keys()])+ 1
    
    return base_json

def calculate_iou(box_1, box_2):
    poly_1 = Polygon(box_1)
    poly_2 = Polygon(box_2)
    iou = poly_1.intersection(poly_2).area / poly_1.union(poly_2).area
    return iou

def xywh2box(box):
    tl = [box[0] - (box[2]/2), (box[1] - (box[3]/2))]
    tr = [box[0] + (box[2]/2), (box[1] - (box[3]/2))]
    bl = [box[0] - (box[2]/2), (box[1] + (box[3]/2))]
    br = [box[0] + (box[2]/2), (box[1] + (box[3]/2))]
    
    return [tl, tr, br, bl]

def box2xywh(box):
    w = box[1][0] - box[0][0]
    h = box[2][1] - box[1][1]
    x = box[0][0] + (w/2)
    y = box[0][1] + (h/2)
    return [x,y,w,h]

def pad(box, n):
    tl = [box[0][0] - n, box[0][1] - n]
    tr = [box[1][0] + n, box[1][1] - n]
    br = [box[2][0] + n, box[2][1] + n]
    bl = [box[3][0] - n, box[3][1] + n]
    return [tl,tr,br,bl]

box_1 = [[511, 41], [577, 41], [577, 76], [511, 76]]
box_2 = [[544, 59], [610, 59], [610, 94], [544, 94]]

print(calculate_iou(box_1, box_2))
```

```python
class ContextFeatures():
    
    def __init__(self,alt=0,time=0,temp=0):
        self.altitude = alt
        self.timeofday = time
        self.temperature = temp
        
        
class ContextBounds():
    
    def __init__(self, a_b=(0,0), time_b=(0,0), temp_b=(0,0)):
        self.alt_bounds = a_b
        self.time_bounds = time_b
        self.temp_bounds = temp_b


    
class BagOfOperators():
    
    def __init__(self):
        self.initial_frames = []
        # context = (ContextBounds,ChI)
        self.contexts = []
        self.default_integral = None
        
    def print_contexts(self):
        for i, context in enumerate(self.contexts):
            print("\nContext " + str(i) + ": altitude" + str(context[0].alt_bounds) + " temperature" +
                  str(context[0].temp_bounds) + " time" + str(context[0].time_bounds) + str(context[1].fm))
            
        print("\nDefault :" + str(self.default_integral.fm))
    
    def add_training_frame(self, contextfeatures):
        self.initial_frames.append(contextfeatures)
    
    def add_context(self, context):
        self.contexts.append(context)
        
    # source :: ContextFeatures, target :: ContextBounds
    def compare_context(self,source,target):
        if target.alt_bounds[0] <= source.altitude <= target.alt_bounds[1]:
            if target.time_bounds[0] <= source.timeofday <= target.time_bounds[1]:
                if target.temp_bounds[0] <= source.temperature <= target.temp_bounds[1]:
                    return 1
        return 0
    
    def gen_default_integral(self):
        ch = chi.ChoquetIntegral()
        ch.type = 'quad'
        ch.N = 2
        ch.fm = chiu.average_fms([x[1].fm for x in self.contexts])
        self.default_integral = ch
    
    #source :: ContextFeatures, output :: fm_dict
    def get_operator(self,source):
        scores = self.score_features(source)
        totals = coll.Counter(scores)
        largest = max(scores)
        full_matches = 0
        
        if largest == 1:
            full_matches = totals[largest]
        
        if full_matches == 1:
            print("Full match")
            return self.contexts[np.argmax(scores)][1]
        
        elif full_matches > 1:
            print("Multi-match!")
            matched_contexts = []
            for i,score in enumerate(scores):
                if score == 1:
                    matched_contexts.append(i)
       
            return chiu.average_fms([self.contexts[x][1] for x in matched_contexts])

        else:
            print("No matches")
            return self.default_integral
            # no known context is a match
            
        
    # source :: ContextBounds
    def score_features(self, source):
        scores = [self.compare_context(source,target[0]) for target in self.contexts]
        return scores
        
```

```python
class Linker():
    
    def __init__(self, bag=None):
        self.total_sources = 2
        self.current_fm = chiu.random_node_fm(4)
        self.current_fm.type = 'quad'
        self.class_name = "Explosive_Hazard"
        self.current_fm.N = self.total_sources
        self.bag = bag
        
    def link_frame(self,fnum,sources):
        unzipped_decls = self.unzip_frame(sources,fnum)
        linked_list = self.link(unzipped_decls)
        full_list = self.fill_missing_sources(linked_list)
        return full_list


    def link_and_fuse_frame(self,fnum,sources,metadata):
        unzipped_decls = self.unzip_frame(sources,fnum)
        linked_list = self.link(unzipped_decls)
        full_list = self.fill_missing_sources(linked_list)
        fused_decls = self.fuse_frame(full_list,metadata)
        return fused_decls
    
    # linked_decls :: (decl,source)
    def fuse_frame(self,linked_decls,metadata):
        operator = self.bag.get_operator(ContextFeatures(alt=metadata[0],temp=metadata[1],time=metadata[2]))
        fused_decls = []
        for target in linked_decls:
            confs = [float(x[0]["confidence"]) for x in target]
            fused_conf = operator.chi_quad(confs)
            
            not_empty_boxes = [xywh2box(x[0]["shape"]["data"]) for x in target if x[0]["shape"] != None]
            fused_box = [[0,0],[0,0],[0,0],[0,0]]
            for predicted_box in not_empty_boxes:
                fused_box = [[(x[0] + y[0]), (x[1] + y[1])] for x,y in zip(predicted_box,fused_box)]   
            fused_box = [[x[0]/len(not_empty_boxes),x[1]/len(not_empty_boxes)] for x in fused_box]
            
            new_decl = {}
            new_decl["confidence"] = fused_conf
            new_decl["shape"] = {}
            new_decl["shape"]["data"] = box2xywh(fused_box)
            new_decl["class"] = self.class_name
            new_decl["shape"]["type"] = "bbox_xywh"
            fused_decls.append(new_decl)
            
        return fused_decls
    
    def create_zero_detection(self):
        linked_decl = []
        
        for i in range(self.total_sources):
            new_decl = {}
            new_decl["confidence"] = 0
            new_decl["class"] = "Explosive_Hazard"
            new_decl["shape"] = None
#             new_decl["shape"["type"] = "bbox_xywh"
            linked_decl.append((new_decl,i))
        
        return linked_decl
        

    def fill_missing_sources(self, linked_decls):
        
        linked_decls_temp = linked_decls.copy()
        for j,target in enumerate(linked_decls_temp):
            if len(target) < self.total_sources:
                for i in range(0,self.total_sources):
                    if i not in [x[1] for x in target]:
                        new_decl = {}
                        new_decl["confidence"] = 0
                        new_decl["class"] = "Explosive_Hazard"
                        new_decl["shape"] = None
#                         new_decl["type"] = "bbox_xywh"
                        linked_decls[j].append((new_decl,i))
                        
            linked_decls[j].sort(key=(lambda x : x[1]))
            
        return linked_decls
    
    #decls :: [(fanno,source_num)]
    def link(self, decls, iou_thresh=.3):
        linked_list = []
        unique = 0
        while len(decls) > 0:
            
            (decl, source) = decls.pop(0)
            linked_list.append([(decl,source)])
            decl_bbox = xywh2box(decl["shape"]["data"])
            decls_temp = decls.copy()
            
            indices = []
            
            for i,(unlinked,unlinked_source) in enumerate(decls_temp):
                target_bbox = xywh2box(unlinked["shape"]["data"])
                
                if source != unlinked_source and calculate_iou(decl_bbox,target_bbox) > iou_thresh:
                    linked_list[unique].append((unlinked,unlinked_source))
#                     decls.pop(i)
                    indices.append(i)

            for index in sorted(indices, reverse=True):
                del decls[index]
                    
            unique = unique + 1
        return linked_list
       
    #annos :: [(fanno,source_num)]
    # decls :: json
    def unzip_frame(self,decls, fnum):
        finished_list = []
        for i,decl_file in enumerate(decls):
            if "f" + str(fnum) in decl_file["frameDeclarations"]:
                for declaration in decl_file["frameDeclarations"]["f" + str(fnum)]["declarations"]:
                    finished_list.append((declaration,i))    
        finished_list.sort(key=(lambda x : x[0]["shape"]["data"][2] * x[0]["shape"]["data"][3]))
         
        return finished_list
            
        
```

```python
class ContextFuser():
    
    def __init__(self):
        self.num_clusters = 0
        self.train_data = None
        self.centers = None
        self.bag = BagOfOperators()
        self.jsons = None
        self.truth = None
        
    def predict_new_data(self, data):
        u, u0, d, jm, p, fpc = fuzz.cluster.cmeans_predict(data, self.centers, 1.1, error=0.005, maxiter=10, init=None, seed=None)
        return u
    
    # len(sources) == len(data)
    # data :: np.array((num features, num samples))
    def cluster_metadata(self,data,max_clusters=6):
        self.train_data = data
        max_fpc = 0
        final_nclusters = 0
        final_centers = 0
        final_u = 0
        
        for nclusters in range(2,min(data.shape[1]+1,max_clusters+1)):
            cntr, u, u0, d, jm, p, fpc = fuzz.cluster.cmeans(data, nclusters, 2, error=0.005, maxiter=1000, init=None)
            if fpc > max_fpc:
                max_fpc = fpc
                final_nclusters = nclusters
                final_centers = cntr
                final_u = u
                self.centers = cntr
            print(str(nclusters) + " clusters: " + str(fpc))
                
        cluster_membership = np.argmax(final_u, axis=0)
        
        
        
        # bounds[cluster num][feature num] = [min, max]
        bounds = []
        for j in range(final_nclusters):
            bound = ContextBounds()
            bound.alt_bounds = [np.min(data[0][cluster_membership == j]), 
                                           np.max(data[0][cluster_membership == j])]
            bound.temp_bounds = [np.min(data[1][cluster_membership == j]), 
                                           np.max(data[1][cluster_membership == j])]
            bound.time_bounds = [np.min(data[2][cluster_membership == j]), 
                                           np.max(data[2][cluster_membership == j])]   
            bounds.append(bound)
            
            relevant_data = [i for i,n in enumerate(cluster_membership) if n == j]
            linked_annotations = self.link_training_data(relevant_data,self.jsons)
            (d,l) = self.massage_linked_data(linked_annotations)
            fm = self.train_massaged_data(d.T,l)
            
            self.bag.add_context((bound,fm))
            
        linked_annotations = self.link_training_data([i for i,n in enumerate(cluster_membership)],self.jsons)
        (g_d,g_l) = self.massage_linked_data(linked_annotations)
        g_fm = self.train_massaged_data(g_d.T,g_l)
        self.bag.gen_default_integral()
            
        return final_nclusters, final_centers.T, bounds

    # relevant frame :: [frame num], jsons :: [json], truth :: json
    def link_training_data(self,relevant_data,jsons):
        l = Linker()
        linked_data = []

        print("len of relevant data:", len(relevant_data))
        for f_num in relevant_data:
#                 if (s_num == source):
            linked_decls = l.link_frame(f_num,jsons)
            linked_data.append((self.link_annotations_to_decls(f_num,linked_decls,),f_num))
        

        return linked_data
            
    
    def link_annotations_to_decls(self,f_num,linked_declus,iou_thresh=.3):
        l = Linker()
        linked_annotations= []
        annotations = self.truth["frameAnnotations"]["f" + str(f_num)]["annotations"]
        linked_decls = linked_declus.copy()
        unique = 0
#         print(linked_decls)
        
        while len(annotations) > 0:
            anno = annotations.pop(0)
            linked_annotations.append((anno,[]))
            anno_bbox = xywh2box(anno["shape"]["data"])
            padded_anno_bbox = pad(anno_bbox,5)
            annotations_temp = annotations.copy()
            # link all annotations to existing declarations

            linked_decls_temp = linked_decls.copy()
            indices = []
            for j,target in enumerate(linked_decls_temp):
                for i, (decl,source) in enumerate(target):
                    if decl["shape"] != None:
                        decl_bbox = xywh2box(decl["shape"]["data"])

                        if calculate_iou(decl_bbox,padded_anno_bbox) > iou_thresh:
                            linked_annotations[unique] = (anno,target)
                            indices.append(j)
                            break
             
            for index in sorted(indices, reverse=True):
                del linked_decls[index]

            if len(linked_annotations[unique][1]) == 0:
                linked_annotations[unique] = (anno,l.create_zero_detection())
                
            unique = unique+1

        # for all excess declarations, they must be false alarm. make an empty annotation
        linked_decls_temp = linked_decls
        for i,target in enumerate(linked_decls_temp):
            linked_annotations.append((None,target))
                        
        return linked_annotations
    
    def massage_linked_data(self,linked_annotations):
        # linked annotations:: [0]['frame'][anno_num] = (anno,decls
        data = np.zeros(len(linked_annotations[0][0][0][1]))
        labels = np.zeros(1)
        
        for (d,frame) in linked_annotations:
            for (anno,decls) in d:
                if anno == None:
                    labels = np.vstack((labels,0))
                else:
                    labels = np.vstack((labels,1))
                
#                 print(decls)
                data = np.vstack((data,np.array([x[0]["confidence"] for x in decls])))

#                 (anno,decls) = d

        data = np.delete(data,0,axis=0)
        labels = np.delete(labels,0,axis=0)
        return (data, labels)
    
    # metadata :: array(3,num_frames)
    def fuse_new_declarations(self, jsons, metadata):
        l = Linker(bag=self.bag)
        new_json = {}
        new_json["source"] = "yolo fusion"
        new_json["frameDeclarations"] = {}
        for i in range(700):
            new_json["frameDeclarations"]["f" + str(i)] = {}
            new_json["frameDeclarations"]["f" + str(i)]["declarations"] = l.link_and_fuse_frame(i,jsons,metadata[:,i])
            
        return new_json
            

            

    def train_massaged_data(self, data,labels):
        chI = chi.ChoquetIntegral()
        chI.train_chi(data,labels)
        return chI
            
```

```python
def gen_metadata(n=700):
    altitudes = np.zeros(1)
    temps = np.zeros(1)
    t_sin = np.zeros(1)
    t_cos = np.zeros(1)
    frames = np.zeros(1)
    sources = np.zeros(1)
    source = 0 

    for i in range(0,n):
        altitudes = np.hstack((altitudes, np.random.normal(10,.1)))
        temps = np.hstack((temps, np.random.normal(70,.3)))
#         time = np.hstack((time,np.cos((2 * np.pi * 0) / 24)))
        t_cos = np.hstack(t_cos,np.cos((2 * np.pi * (i/n)) / 23))
        t_sin = np.hstack(t_sin,np.sin((2 * np.pi * (i/n)) / 23))
        frames = np.hstack((frames,i))
        sources = np.hstack((sources,source))



    source = 1
    for i in range(0,n):
        altitudes = np.hstack((altitudes, np.random.normal(10,.1)))
        temps = np.hstack((temps, np.random.normal(70,.3)))
        t_cos = np.hstack(t_cos,np.cos((2 * np.pi * (6+(i/n))) / 23))
        t_sin = np.hstack(t_sin,np.sin((2 * np.pi * (6+(i/n))) / 23))
        frames = np.hstack((frames,i))
        sources = np.hstack((sources,source))

    all_data = np.vstack((altitudes,temps,t_cos,t_sin))
    frames = np.delete(frames,0)
    sources = np.delete(sources,0)
    all_data = np.delete(all_data,0, axis=1)
    return all_data, frames, sources
```
